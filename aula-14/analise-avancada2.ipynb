```python
# Metodo apply  (criando faixa etaria)


import numpy as np
```

```python
def calc_faixa_etaria(idade):

    if np.isnan(idade):
        return "desconhecida"
    
    elif idade <= 15:
        return "0-15"

    elif idade <= 30:
        return "15-30"

    elif idade <= 45:
        return "30-45"

    else:
        return "45+"
```

```python
df["age_group"] = df["Age"].apply(calc_faixa_etaria)
```

```python
df
```

```python
df["age_group"].value_counts()
```

```python
df.groupby(["age_group", "Sex"])[["Sex"]].count()
```

```python
# taxa de morte/sobrevivencia por faixa etaria

df.groupby(["age_group"])[["Survived"]].agg(["mean", "count"])
```

```python
df.groupby(["age_group", "Sex"])[["Survived"]].agg(["mean", "count"])
```
