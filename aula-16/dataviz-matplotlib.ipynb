```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# ponto x do intervalo a ser plotado

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# função que faz o plot dos pontos

plt.plot(x, y, marker="o")
```

```python
# array com 20 pontos igualmente espaçados entre 1 e 5

x = np.linspace(1, 5, 20)
y = x**2

plt.plot(x, y, marker="o")
```

```python
# y = sqrt(x), de 0 a 10  

x = np.linspace(0, 10, 1000)

plt.plot(x, np.sqrt(x))
```

```python
# y = e^x, de -5 a 5

x = np.linspace(-5, 5, 1000)

plt.plot(x, np.exp(x))
```

```python
# inicializando uma figura com o tamanho alterado

plt.figure(figsize=(8, 4))

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

```python
# salvando a figura

figure = plt.figure(figsize=(8, 4))

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()

figure.savefig("./figura1.png")
```
