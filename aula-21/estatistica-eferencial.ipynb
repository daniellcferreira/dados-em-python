```python
# estatistica eferencial

# teorema central do limite - tcl

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import scipy
```

```python
# distribuição uniforme

a, b = 0, 10

plt.title("Distribuição Populacional Uniforme")
x = np.linspace(a, b, 100000)
y = np.ones(100000)/(b-a)

plt.plot(x, y)

plt.axvline(x=a, ymax=0.5)
plt.axvline(x=b, ymax=0.5)

plt.show()
```

```python
n = 30
n_amostras = 100000

medias_amostrais = []

for _ in range(n_amostras):

    amostra = np.random.uniform(low=a, high=b, size=n)
    media_amostral = amostra.mean()

    medias_amostrais.append(media_amostral)

#print(f"amostra: {amostra}\nmedia: {media_amostral}")
```

```python
plt.title("Distribuição amostrais")
sns.histplot(medias_amostrais)

plt.show()
```

```python
np.array(medias_amostrais).std()
```

```python
# distribuição gamma

import scipy.special


a, b = 2, 1/2

plt.title("Distribuição Populacional Gamma")
x = np.linspace(0, 30, 100000)
y = ((b**a)*(x**(a-1))*np.exp(-b*x))/scipy.special.gamma(a)

plt.plot(x, y)

plt.show()
```

```python
n = 50
n_amostras = 100000

medias_amostrais = []

for _ in range(n_amostras):

    amostra = np.random.gamma(shape=a, scale=1/b, size=n)
    media_amostral = amostra.mean()

    medias_amostrais.append(media_amostral)
```

```python
plt.title("Distribuição amostrais")
sns.histplot(medias_amostrais)

plt.show()
```

```python
# destribuição exponencial

l = 0.5

plt.title("Distribuição Populacional Exponencial")
x = np.linspace(0, 30, 100000)
y = l*np.exp(-l*x)

plt.plot(x, y)

plt.show()
```

```python
n = 50
n_amostras = 100000

medias_amostrais = []

for _ in range(n_amostras):

    amostra = np.random.exponential(scale=1/l, size=n)
    media_amostral = amostra.mean()

    medias_amostrais.append(media_amostral)
```

```python
plt.title("Distribuição amostrais")
sns.histplot(medias_amostrais)

plt.show()
```
