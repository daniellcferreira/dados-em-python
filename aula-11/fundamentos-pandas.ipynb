```python
# criando nova coluna

import random

df["Prova_5"] = [random.randint(0, 10) for _ in range(10)]

df
```

```python
# criando nova linha

df.loc[10, "RA"] = 110237
df.loc[10, "Nome"] = "Han Solo"
df.loc[10, "Frequencia"] = 25
df.loc[10, "Prova_1"] = 10
df.loc[10, "Prova_2"] = 8
df.loc[10, "Prova_3"] = 5
df.loc[10, "Prova_4"] = 6.5
df.loc[10, "Prova_5"] = 7
```

```python
df
```

```python
# excluindo uma coluna

df = df.drop(columns=["Prova_5"])
```

```python
df
```

```python
# excluindo um linha

df = df.drop(index=[10])
```

```python
df
```

```python
# operações

df["Media"] = df[["Prova_1", "Prova_2", "Prova_3", "Prova_4"]].mean(axis=1)
```

```python
df
```

```python
# Filtrando dados

df[df["Media"] > 8]
```

```python
df[(df["Media"] > 7) & (df["Frequencia"] < 19)]
```

```python
df[(df["Media"] > 7) | (df["Frequencia"] < 19)]
```

```python
# Filtros com query

df.query("Media > 8 and Frequencia < 20")
```

```python
# salvando arquivo

df.to_csv("notas_processadas.csv", sep=";", index=False)
```
